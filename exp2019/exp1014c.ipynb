중요 날짜 이벤트 annotation

```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy'),
]
```

```python
fig, ax = plt.subplots()

spx.plot(ax=ax, style='k--')

for date, event in crisis_data:
    ax.annotate(event, xy=(date, spx.asof(date) + 75), 
                xytext=(date, spx.asof(date) + 225), 
                arrowprops=dict(facecolor='red', headwidth=4, width=2, headlength=4), 
                horizontalalignment='left', verticalalignment='top')

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris_dataset = datasets.load_iris()
```

```python
print("iris_dataset key:{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['target'])
print(iris_dataset['target_names'])
```

```python
print(iris_dataset['DESCR'])
```

다양한 머신러닝 분류기법

k-nearest neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# 모든 feature 사용

X = iris_dataset.data
y = iris_dataset.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=35)
```

```python
np.bincount(y_train)
```

```python
# 모델 객체 생성
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# 모델 훈련
clf.fit(X_train, y_train)
```

```python
# 예측
predict_label = clf.predict(X_test)
```

```python
predict_label == y_test
```

```python
print('test accuracy {:.2f}'.format(np.mean(predict_label == y_test)))
```

```python
plt.scatter(X_train[:,0], X_train[:,1], edgecolors='k', s=20, c=y_train)
```

```python
plt.scatter(X_train[:,2], X_train[:,3], edgecolors='k', s=20, c=y_train)
```

```python
print("트레인 세트 정확도: {:.2f}".format(clf.score(X_train, y_train)))
```

```python
print("테스트 세트 정확도: {:.2f}".format(clf.score(X_test, y_test)))
```

(!) 위에서 시각화한 아래 그래프의 레이블 레전드를 표시해보기

```python
plt.scatter(X_train[:,2], X_train[:,3], edgecolors='k', s=20, c=y_train)
```

```python
mask0 = y_train == 0
mask1 = y_train == 1
mask2 = y_train == 2

plt.scatter(X_train[mask0,2], X_train[mask0,3], edgecolors='k', s=20, c='r', label='class0')
plt.scatter(X_train[mask1,2], X_train[mask1,3], edgecolors='k', s=20, c='g', label='class1')
plt.scatter(X_train[mask2,2], X_train[mask2,3], edgecolors='k', s=20, c='b', label='class2')

plt.legend(title='classes')
```

```python
fig, ax = plt.subplots()
colors = ['r','g','b']
classes = [0,1,2]

for i, color in zip(classes, colors):
    idx = y_train == i
    ax.scatter(X_train[idx,2], X_train[idx,3], edgecolor='k', c=color, label='class{}'.format(i))
    ax.legend(loc='center left', title='classes')
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(X_train[:,2], X_train[:,3], edgecolor='k', c=y_train)

ax.legend(*scatter.legend_elements(), loc='lower right', title='classes')
```
